```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("ecommerce_customer_data_large.csv")

df["Purchase Date"] = pd.to_datetime(df["Purchase Date"])
```

```python
print("Total Customers:",df["Customer ID"].nunique())
```

```python
print("Total Transactions:",len(df))
```

```python
df["Churn"].value_counts()
```

```python
df["Churn"].value_counts()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(data=df,x="Gender")

plt.title("Gender Distribution")

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.countplot(data=df,x="Product Category")

plt.xticks(rotation=45)

plt.title("Product Categories")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.histplot(df["Customer Age"],bins=20,kde=True)

plt.title("Age Distribution")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.boxplot(x=df["Total Purchase Amount"])

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.countplot(data=df,x="Payment Method")

plt.xticks(rotation=30)

plt.show()
```

```python
import numpy as np
numeric = df.select_dtypes(include=np.number)

plt.figure(figsize=(8,6))

sns.heatmap(numeric.corr(),annot=True,cmap="Blues")

plt.show()
```

```python
df["Month"] = df["Purchase Date"].dt.month

monthly = df.groupby("Month")["Total Purchase Amount"].sum()

monthly.plot(kind="line",marker="o")

plt.title("Monthly Sales")

plt.show()
```
